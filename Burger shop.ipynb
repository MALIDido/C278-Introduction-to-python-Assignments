```python
# implement the classes listed below 
class FoodItem:
    def __init__(self):
        self.beaf = 5
        self.chicken = 4
        self.onion = 0.5
        self.tomato = 0.5
        self.chili = 0.25
        self.fries = 2
        self.onion_ring = 1.5
        self.salad = 1.5
        self.water = 1
        self.soda_s=1.5
        self.soda_m=2
        self.soda_l=2.5
class Burger(FoodItem):
    def __init__(self,beaf_or_chicken,want_onion,want_tomato,chili_or_normal):
    
        self.beaf_or_chicken= beaf_or_chicken
        self.want_onion=want_onion
        self.want_tomato=want_tomato
        self.chili_or_normal=chili_or_normal
        self.total_price=0


    def get_data_price(self):
        burger_price= dict()
        super().__init__()
        
        burger_price["type"]="burger"
        if self.beaf_or_chicken==1:
            burger_price["beaf"]=self.beaf
            self.total_price+=self.beaf
            
        else:

            burger_price["chicken"]=self.chicken
            self.total_price+=self.chicken
                        
        if self.want_onion==1:
            burger_price["onion"]=self.onion
            self.total_price+=self.onion

            
        if self.want_tomato==1:
            burger_price["tomato"]=self.tomato
            self.total_price+=self.tomato

            
        if self.chili_or_normal==1:
            burger_price["chili"]=self.chili

        return burger_price, self.total_price
            

            
class Drink(FoodItem):
    def __init__(self,drink_type):
    
        self.drink_type= drink_type

        self.total_price=0


    def get_data_price(self):
        drink_price= dict()
        super().__init__()

        drink_price["type"]="drink"
        if self.drink_type==1:
            drink_price["water"]=self.water
            self.total_price+=self.water
            
        elif self.drink_type==2:
            drink_price["soda_s"]=self.soda_s
            self.total_price+=self.soda_s
            
            
        elif self.drink_type==3:
            drink_price["soda_m"]=self.soda_m
            self.total_price+=self.soda_m

        else:
            drink_price["soda_l"]=self.soda_l
            self.total_price+=self.soda_l

        return drink_price,self.total_price

    
class Side(FoodItem):
    def __init__(self,side_type):
    
        self.side_type= side_type

        self.total_price=0


    def get_data_price(self):
        side_price= dict()
        super().__init__()

        side_price["type"]="side"
        if self.side_type==1:
            side_price["fries"]=self.fries
            self.total_price+=self.fries
            
        elif self.side_type==2:
            side_price["onion_ring"]=self.onion_ring
            self.total_price+=self.onion_ring
            

        else:
            side_price["salad"]=self.salad
            self.total_price+=self.salad

        return side_price,self.total_price

    
class Combo(FoodItem):
    def __init__(self):
        self.my_burger = dict()
        self.total_price= 0
    def get_data_price(self):
        burger_object=user_input_burger()
        
        burger_item_list,price= burger_object.get_data_price()
        
        self.total_price+=price
        
        
        drink_object=user_input_drink()
        
        drink_item_list,price= drink_object.get_data_price()
        
        self.total_price+=price
        
        
        side_object=user_input_side()
        
        side_item_list,price= side_object.get_data_price()
        
        self.total_price+=price
        
        
        
        self.my_burger["Combo"]=[burger_item_list,drink_item_list,side_item_list]
        
        return self.my_burger, (self.total_price-(self.total_price*0.15))
        
    
class Order:
    def __init__(self):
        self.my_order = list()
#         self.price= 0
        self.total_price= 0

    def get_burger(self):
        ob=user_input_burger()
        item_list,price= ob.get_data_price()
        
        self.total_price+=price

        self.my_order.append(item_list)
        
    def get_drink(self):
        ob=user_input_drink()
        item_list,price= ob.get_data_price()
        
        self.total_price+=price

        self.my_order.append(item_list)
        
    def get_side(self):
        ob=user_input_side()
        item_list,price= ob.get_data_price()
        
        self.total_price+=price

        self.my_order.append(item_list)
        
        
    def get_combo(self):
        ob=user_input_combo()
        item_list,price= ob.get_data_price()
        
        self.total_price+=price

        self.my_order.append(item_list)
    
    def display(self):
        new_combo= dict()
        for i in self.my_order:
            for j in i:
                if j=="Combo":
                    print ("type", j)
#                     new_combo= self.my_order[j][0]
                    for k in i[j]:
                        for l in k:
                            print (l, k[l])
                        print("**************************************************")
                else:
                    print (j,i[j])
            print("==========================================================")
        print("Total price = $", self.total_price)

    
        
         
def user_input_burger():
#     b = Burger()
    #ask user for input and store it in burger object 
    print("burger choice")
    burger_type=int(input("[1]beaf or [2]chicken"))
    tomato=int(input("whould you like tomato on that [1]yes [2]no"))
    onion=int(input("whould you like onion on that [1]yes [2]no"))
    spicy=int(input("whould you like it spicy [1]yes [2]no"))
    b = Burger(burger_type,tomato,onion,spicy)
    return b
 
def user_input_drink():
#     d = Drink()
    #ask user for input and store it in drink object
    print("drink choice")
    drink_type=int(input("[1]Water [2]small soda [3]medium soda [4] large soda"))
    d = Drink(drink_type)
    return d
 
def user_input_side():
#     s = Side()
    #ask user for input and store it in side object 
    print("side choice")
    side_type=int(input("[1]fries [2] onion rings [3]salad"))
    s = Side(side_type)
    return s
 
def user_input_combo():
    c = Combo()
    print("combo choice")
    #ask user for input and store it in combo object 
    #a combo must include one burger, one side, and one drink
    return c
 
def take_order():
    #ask user for name for the order 
    Name= input("Can I have your name please for the order?")
    print("Here is the menu")
    print("Burger:\n\tBeaf= $5     Chicken= $4\n\tTomato= $0.5     Onion= $0.5     spicy=$0.25")
    print("Sides:\n\tFries= $2     onion rings= $1.5     salad= $1.5")
    print("Drinks:\n\tWater= $1\n\tSoda     S= $1.5     M= $2     L= $2.5")
    print("Combo:\n\tBurger+Side+Drink (15% off from the total of a combo)")
    flag=True
    bb=Order()
#     burger
    while flag:
        choice= int(input("please choose: [1] Burger [2] Side [3] Drink [4] Combo [5]Done [6]Cancle the order"))
        if choice==6:
            print("Sorry for that, come again")
            flag=False
        elif choice==1:
            bb.get_burger()
        elif choice==2:
            bb.get_side()
        elif choice==3:
            bb.get_drink()
        elif choice==4:
            bb.get_combo()
        elif choice==5:
            print("customer name: ", Name)
            bb.display()
            print("Thank you for your order")
            flag=False
        else:
            print("invalid choice try again")
    
    #repeat taking order until client is done 
    #display order details
    #display a thank you message
    
    print("Welcome to Burger Shop")
 
take_order()
```
